# Example of Chi-Square and Fisher Score Feature Selection

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
```

```python
iris= load_iris()
X = iris.data
y = iris.target

X = X.astype(int)
```

# Chi-Square method

```python
selector = SelectKBest(chi2, k=2) # number of features to select "k"
newX = selector.fit_transform(X, y)
```

```python
print('Original number of features:', X.shape[1])
print('Reduced number of features:', newX.shape[1])
```

# Fisher-Score method

```python
from skfeature.function.similarity_based import fisher_score
import matplotlib.pyplot as plt
```

```python
ranks = fisher_score.fisher_score(X, y)
plt.xticks(rotation=330)
plt.bar(iris.feature_names, ranks)
plt.show()
```

Here we select those features which have high fisher score

# Information Gain Method

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
importance = mutual_info_classif(X, y)
plt.xticks(rotation=330)
plt.bar(iris.feature_names, importance)
plt.show()
```

Here we select those features which have high information gain in context of target feature
